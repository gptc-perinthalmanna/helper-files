```python
import csv
import json

data = {}


f = open('Dip R(15) O 2021 (2)-01-00011.csv', 'r')

# create the csv writer
reader = csv.reader(f)
next(reader)

for item in reader:
    if not item[0] in data:
        data[item[0]] = {
            "Branch": item[2],
            "Name": item[1],
            "Semester": item[3],
        }
    if item[9] == "F" and item[6] == 'Absent':
        data[item[0]][item[4]] = 'Absent'
    else:
        data[item[0]][item[4]] = item[9]
f.close()
print(json.dumps(data, indent=4))
```

```python
f = open('Sorted_S3_Result-November-2020-Grades.csv', 'w')
writer = csv.writer(f)

# subjects = ['6045-Radar and Navigation',
# '6049-Computer Hardware and Networking Lab',
# '6048-HDL and Simulation Software Lab',
# '6043-Computer Hardware & Networking',
# '6042-Communication Systems',
# '6041-Advanced Microprocessors',
# '6009-Project & Seminar']

# subjects = ['1001-English for Communication I',
# '1002-Engineering Mathematics I',
# '1003-Engineering Physics I',
# '1004-Engineering Chemistry I',
# '1008-Computing Fundamentals',
# '1009-Health & Physical Education']

subjects = ['3042-Digital Electronics',
'3043-Electrical Technology',
'3044-Electronic Devices & Circuits',
'3047-Electronics Circuits Lab',
'3048-Digital Electronics Lab',
'3049-Analogue Communication Lab',
'3001-Environmental Science & Disaster Management',
'3041-Communication Engineering']


writer.writerow(['RegNo', 'Name', 'Branch', 'Semester'
]+ subjects)

for key in data:
   
    row = []
    row.append(key)
    row.append(data[key]['Name'])
    row.append(data[key]['Branch'])  
    row.append(data[key]['Semester'])
    for sub in subjects:
        if sub in data[key]: 
            row.append(data[key][sub])
        else:
            row.append('Not Found')
    print(row)
    writer.writerow(row)
```

```python
# Calculate total status

status = {}

for sub in subjects:
    status[sub] = {}
for key in data:
    val = data[key]
    for sub in subjects:
        if sub in val:
        
            if val[sub] in status[sub]:
                status[sub][val[sub]] += 1
            else:
                status[sub][val[sub]] = 1

            
print(json.dumps(status, indent=4))
```

```python
f = open('Grade-status-S3.csv', 'w')
writer = csv.writer(f)

subjects = ['S', 'A', 'B', 'C', 'D', 'E', 'F', 'Absent']

writer.writerow(['Subject',
]+ subjects)

for key in status:
    row = []
    row.append(key)
    for grade in subjects:
        if grade in status[key]:
            row.append(status[key][grade])
        else:
            row.append(0)
    print(row)
    writer.writerow(row)
```
